```python
n = int(input("Enter Your Number: "))
for num in range(1, n + 1):        # Table Loop 1 to n
    print(f"\nTable of {num}")
    total = 0
    for i in range(1, 11):        # 1 to 10 table multiply
        result = num * i
        print(f"{num} x {i} = {result}")
        total += result
    print(f"Total Sum of Table {num} = {total}")
```

```python
items = int(input("How many items you want to purchase? "))
customer_name = input("Enter Customer Name: ")
cart = []  
total = 0
print("\n------ Bill Details ------")
print("\ncustomer_name",customer_name)
for i in range(1, items + 1):
    name = input(f"Enter Item {i} Name: ")
    price = float(input("Enter Price: "))
    qty = int(input("Enter Quantity: "))   
    amount = price * qty
    total += amount
    cart.append({
        "Item": name,
        "Price": price,
        "Qty": qty,
        "Amount": amount
    })
    print(f"{name}  x {qty}  = ₹{amount}")
gst = total * 0.18
final_amount = total + gst

print("Total Purchase Amount:",total)
print("Totak GST:",gst)
print("Total Bill Amount:",final_amount)
```
