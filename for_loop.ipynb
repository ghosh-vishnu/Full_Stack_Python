```python
print("=================WELCOME TO SBI ATM==================")
set_pin=int(input("Enter Your PIN:"))
balance=10000
password=set_pin
attempt=3
while attempt>0:
    login=int(input("Enter Your PIN"))
    
if password==login:
    print("\n==============Login Successfully===============\n")
    
    print("========ATM MENU==========")
    print("1.Check Balance")
    print("2.Withdraw Money")
    print("3.Deposit Money")
    print("4.Pin Change")
    
    choose=int(input("Enter Your Choise"))
    if choose==1:
        print("Your Balance is:",balance)
    elif choose==2:
        withdraw_amount=int(input("Enter Your Amount:"))
        if withdraw_amount<=10000 and withdraw_amount>=1:
            if withdraw_amount % 100==0:
                print("Withdraw SuccessFully")
                balance-=withdraw_amount
                print("-------------------------")
                print("Total Withdraw Amount:",withdraw_amount)
                print("Total Remaining Amount:",balance)
            else:
                print("Please Enter Multiplese of 100")
        else:
            print("Incoorect Amount")
            print("Insufficient Amount")
    elif choose==3:
        deposit=int(input("Enter Your Deposit Amount:"))
        if deposit>=0:
            print("Deposit Successfully")
            balance+=deposit
        else:
            print("Please Enter Valid Amount")
    elif choose==4:
        update_pin=int(input("Enter Your NEW PIN"))
        password=update_pin
else:
    print("Incorrect Password")

print("Thank You")
```
