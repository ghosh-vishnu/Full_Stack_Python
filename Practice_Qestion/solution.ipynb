```python
print("=================WELCOME TO SBI ATM==================")

set_pin = int(input("Set Your PIN: "))
balance = 10000
password = set_pin
attempt = 3

while attempt > 0:
    login = int(input("Enter Your PIN: "))

    if password == login:
        print("\n============== Login Successfully ===============\n")

        print("======== ATM MENU ========")
        print("1. Check Balance")
        print("2. Withdraw Money")
        print("3. Deposit Money")
        print("4. Pin Change")

        choose = int(input("Enter Your Choice: "))

        if choose == 1:
            print("Your Balance is:", balance)

        elif choose == 2:
            withdraw_amount = int(input("Enter Amount to Withdraw: "))
            if withdraw_amount <= balance and withdraw_amount >= 1:
                if withdraw_amount % 100 == 0:
                    print("Withdraw Successful!")
                    balance -= withdraw_amount
                    print("-------------------------")
                    print("Withdraw Amount:", withdraw_amount)
                    print("Remaining Balance:", balance)
                else:
                    print("Please Enter Amount in Multiple of 100 Only")
            else:
                print("Insufficient Balance or Invalid Amount")

        elif choose == 3:
            deposit = int(input("Enter Amount to Deposit: "))
            if deposit > 0:
                print("Deposit Successful!")
                balance += deposit
                print("Updated Balance:", balance)
            else:
                print("Please Enter Valid Amount")

        elif choose == 4:
            update_pin = int(input("Enter Your NEW PIN: "))
            password = update_pin
            print("PIN Updated Successfully!")

        print("\nThank You!")
        break

    else:
        attempt -= 1
        print("Incorrect PIN! Attempts Left:", attempt)

        if attempt == 0:
            print("\n Your ATM Card Has Been Blocked For Security Reasons")
```

```python
from datetime import datetime, timedelta

print("=================WELCOME TO MULTI-PURPOSE SYSTEM==================")

balance = 10000
attempt = 3
password = None
block_time = None

# Set 4-digit PIN
while True:
    set_pin = input("Set Your 4-Digit ATM PIN: ")
    if set_pin.isdigit() and len(set_pin) == 4:
        password = int(set_pin)
        break
    else:
        print("Invalid PIN — must be exactly 4 digits\n")

# -------- MAIN MENU LOOP --------
while True:
    print("\n===== MAIN MENU =====")
    print("1. ATM")
    print("2. Electricity Bill Calculator")
    print("3. Train Ticket Booking")
    print("4. School Admission Age Check")
    print("5. Water Bottle Discount Scheme")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        # -------- ATM MODULE --------
        if block_time is not None:
            current_time = datetime.now()
            if current_time < block_time:
                formatted = block_time.strftime("%d-%m-%Y %I:%M %p")
                print("\nYour ATM Card is BLOCKED ")
                print("You can try again after:", formatted)
                continue
            else:
                block_time = None
                attempt = 3
                print("\n✔ Card has been Unblocked! You can Access Your Account Now.\n")

        login = int(input("Enter Your PIN: "))

        if login == password:
            print("\n============== Login Successfully ===============\n")
            print("-------- ATM MENU --------")
            print("1. Check Balance")
            print("2. Withdraw Money")
            print("3. Deposit Money")
            print("4. PIN Change")

            atm_choice = input("Enter your choice: ")

            if atm_choice == "1":
                print("Your Balance is:", balance)

            elif atm_choice == "2":
                withdraw_amount = int(input("Enter Amount to Withdraw: "))
                if 1 <= withdraw_amount <= balance:
                    if withdraw_amount % 100 == 0:
                        balance -= withdraw_amount
                        print("Withdraw Successful! Remaining Balance:", balance)
                    else:
                        print("Please enter amount in multiples of 100 only")
                else:
                    print("Insufficient Balance or Invalid Amount")

            elif atm_choice == "3":
                deposit = int(input("Enter Amount to Deposit: "))
                if deposit > 0:
                    balance += deposit
                    print("Deposit Successful! Updated Balance:", balance)
                else:
                    print("Enter a valid amount")

            elif atm_choice == "4":
                while True:
                    new_pin = input("Enter New 4-Digit PIN: ")
                    if new_pin.isdigit() and len(new_pin) == 4:
                        password = int(new_pin)
                        print("PIN Updated Successfully!")
                        break
                    else:
                        print("Invalid PIN! Must be 4 digits.")

        else:
            attempt -= 1
            print("Incorrect PIN! Attempts Left:", attempt)
            if attempt == 0:
                block_time = datetime.now() + timedelta(hours=24)
                formatted = block_time.strftime("%d-%m-%Y %I:%M %p")
                print("\nCard BLOCKED due to 3 Wrong Attempts")
                print("You can try again after:", formatted)

    elif choice == "2":
        # -------- ELECTRICITY BILL --------
        units = int(input("Enter total units consumed: "))
        bill = 0
        if units <= 100:
            bill = units * 5
        elif units <= 200:
            bill = (100*5) + ((units-100)*7)
        elif units <= 300:
            bill = (100*5) + (100*7) + ((units-200)*10)
        else:
            bill = (100*5) + (100*7) + (100*10) + ((units-300)*15)
        print("Total Units:", units)
        print("Total Electricity Bill: ₹", bill)

    elif choice == "3":
        # -------- TRAIN TICKET BOOKING --------
        print("Available Classes:\n1. Sleeper (₹300)\n2. AC (₹800)\n3. General (₹150)")
        class_choice = input("Choose Class (1/2/3): ")
        tickets = int(input("Number of Tickets: "))

        if class_choice == "1":
            base_price = 300
            class_name = "Sleeper"
        elif class_choice == "2":
            base_price = 800
            class_name = "AC"
        elif class_choice == "3":
            base_price = 150
            class_name = "General"
        else:
            print("Invalid Class Choice")
            continue

        tatkal = input("Tatkal Booking? (yes/no): ").lower()
        extra = 200 * tickets if tatkal == "yes" else 0
        total_price = (base_price * tickets) + extra

        print("\n--- TICKET SUMMARY ---")
        print("Class:", class_name)
        print("Tickets:", tickets)
        print("Base Price per Ticket:", base_price)
        print("Tatkal Charges:", extra)
        print("TOTAL PRICE: ₹", total_price)

    elif choice == "4":
        # -------- SCHOOL ADMISSION --------
        age = float(input("Enter Child's Age: "))
        if age >= 5:
            print("Admission Approved! Eligible for school.")
        elif age > 0:
            print("Admission Rejected! Age less than 5 years.")
        else:
            print("⚠ Invalid Age Entered!")

    elif choice == "5":
        # -------- WATER BOTTLE DISCOUNT --------
        unit_price = float(input("Enter price per bottle: "))
        quantity = int(input("Enter quantity: "))
        subtotal = unit_price * quantity

        if quantity <= 5:
            discount_rate = 0
        elif quantity <= 11:
            discount_rate = 0.05
        else:
            discount_rate = 0.10

        discount_amount = subtotal * discount_rate
        total = subtotal - discount_amount

        print("\n--- BILL SUMMARY ---")
        print("Unit Price:", unit_price)
        print("Quantity:", quantity)
        print("Subtotal: ₹", subtotal)
        print("Discount:", discount_rate*100, "%")
        print("Discount Amount: ₹", discount_amount)
        print("TOTAL PAYABLE: ₹", total)

    elif choice == "6":
        print("Thank you! Exiting system")
        break

    else:
        print("Invalid Choice! Try Again.")
```
