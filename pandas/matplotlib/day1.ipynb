```python
import matplotlib.pyplot as plt
```

```python
x=['A','B','C','D']
Y=[10,20,15,25]
plt.plot(x,Y)
plt.show()
```

```python
plt.plot(x,Y,marker='s',linestyle='--',color='r')
plt.title("Sample Plot")
plt.xlabel("Student Name")
plt.ylabel("Marks Obtained")
plt.grid(linestyle='--',color='blue')
plt.show()
```

```python
y2=[12,18,22,28]
y3=[14,16,19,30]
plt.plot(x,Y,label='2025',marker='o',linestyle='--',color='r',markersize=8,linewidth=2,markerfacecolor='yellow',markeredgewidth=1,markeredgecolor='black')
plt.plot(x,y2,label='2026',marker='s',linestyle='--',color='g',markersize=8,linewidth=2,markerfacecolor='orange',markeredgewidth=1,markeredgecolor='black')
plt.plot(x,y3,label='2027',marker='*',linestyle='--',color='b',markersize=8,linewidth=2,markerfacecolor='pink',markeredgewidth=1,markeredgecolor='black')
plt.title("Sample Plot for 3 years")
plt.legend(shadow=True,loc='upper center') 
plt.show()
```

```python
# Bar Chart
names=['A','B','C','D']
marks=[25,30,15,10]
a=plt.bar(names,marks,color='red',edgecolor='black',width=0.4,hatch='/')
plt.title("Bar Chart Example")
plt.xlabel("Student Name")
plt.bar_label(a,fontsize=8,color='blue',padding=3)
plt.show()
```

```python
names=['A','B','C','D']
marks=[25,30,15,10]
bars=plt.bar(names,marks,color='red',edgecolor='black',width=0.4,hatch='/')
plt.title("Bar Chart Example with Values")
plt.bar_label(bars)
plt.xlabel("Student Name")
plt.show()
```

```python
import matplotlib.pyplot as plt
data={
    'a':[45,75,58,69,52],
    'b':[35,65,80,59,62],
    'c':[25,85,78,99,82],
    'd':[55,95,68,79,72]
}
for key, value in data.items():
    x=range(len(value))
    print(x)
    plt.scatter(x,value,label=key)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,6])
y=np.array([10,20,25,30,40,58])
plt.scatter(x,y,s=100,c='red',marker='*',alpha=0.5)
plt.title("Scatter Plot Example")
plt.show()
```

```python
temp=[30,23,33,54,85,45,63,25]
windspeed=[4,5,8,9,2,6,5,4]
plt.scatter(temp,windspeed,color='blue')
plt.title("Temp Vs Windspeed")
plt.xlabel("Temp(C)")
plt.ylabel("windspeed(%)")
plt.show()
```
