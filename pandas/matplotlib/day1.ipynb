```python
plt.plot(x,Y,marker='s',linestyle='--',color='r')
plt.title("Sample Plot")
plt.xlabel("Student Name")
plt.ylabel("Marks Obtained")
plt.grid(linestyle='--',color='blue')
plt.show()
```

```python
import matplotlib.pyplot as plt

wheater = {
    "temp": [30, 32, 35, 45, 85, 96],
    "windspeed": [33, 30, 25, 28, 25, 45],
    "event": ['Sunny', 'Rainy', 'Cloudy', 'Sunny', 'Rainy', 'Cloudy']
}

# 1 unique categories
events = list(set(wheater["event"]))
print(events)

# 2 category  number
event_to_num = {event: i for i, event in enumerate(events)}
print(event_to_num)
event_codes = [event_to_num[e] for e in wheater["event"]]
print(event_codes)

# 3 scatter with automatic colormap
scatter = plt.scatter(
    wheater["temp"],
    wheater["windspeed"],
    c=event_codes,
    cmap="Dark2_r",   # automatic distinct colors
    s=100
)

# 4 legend manually (IMPORTANT)
plt.legend(
    handles=scatter.legend_elements()[0],labels=events,title="Weather Event"
)

# print(scatter.legend_elements()[2])
plt.title("Temp vs Wind Speed")
plt.xlabel("Temperature")
plt.ylabel("Wind Speed")
plt.show()
```
