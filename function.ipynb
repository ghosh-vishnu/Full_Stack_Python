```python
from datetime import datetime
def check_balance(balance):
    print("\nCurrent Balance:", balance)
def withdraw(balance,history):
    print("\n Withdraw Method")
    print("1. Cash Withdraw")
    print("2. Upi Withdraw")
    option=int(input("Choose Withdraw Method"))
    amount = int(input("Enter Amount to Withdraw: "))
    if amount <= 0:
        print("Invalid Amount")
    elif amount > balance:
        print("Insufficient Balance")
    elif amount % 100 != 0 and option==1:
        print("Amount must be multiple of 100")
    else:
        balance -= amount
        time=datetime.now().strftime("%d-%m-%Y-%H:%M%S")
        history.append(f"Withdaw Amount is {amount} on {time}")
        print("Withdraw Successful")
        print("Remaining Balance:", balance)
        receipt(balance,amount,"withdraw")
    return balance,history


def deposit(balance):
    amount = int(input("Enter Amount to Deposit: "))
    if amount <= 0:
        print("Invalid Deposit Amount")
    else:
        balance += amount
        print("Deposit Successful")
        print("Updated Balance:", balance)
        receipt(balance,amount,"deposit")
    return balance


def change_pin(pin):
    new_pin = int(input("Enter New PIN: "))
    confirm = int(input("Re-enter New PIN to Confirm: "))
    if new_pin == confirm:
        print("PIN Changed Successfully")
        return new_pin
    else:
        print("PIN Not Matched")
        return pin
    
def receipt(balance, amount, trans_type):
    print("\n ---------- RECEIPT ----------")
    print("Transaction Type:", trans_type)
    print("Amount:", amount)
    print("Available Balance:", balance)
    print("Time:", datetime.now().strftime("%d-%m-%Y %H:%M:%S"))
    print("----------------------------------")

pin = 1234
balance = 50000
attempts = 3
history=[]

print("===== Welcome to SBI ATM =====")
while attempts > 0:
    user_pin = int(input("\nEnter PIN: "))
    if user_pin == pin:
        print("\nLogin Successful\n")
        while True:
            print("\n-------- ATM MENU --------")
            print("1. Check Balance")
            print("2. Withdraw Money")
            print("3. Deposit Money")
            print("4. Change PIN")
            print("5. Exit ATM")
            choice = int(input("\nChoose Option: "))

            if choice == 1:
                check_balance(balance)

            elif choice == 2:
                balance,history = withdraw(balance,history)

            elif choice == 3:
                balance = deposit(balance)

            elif choice == 4:
                pin = change_pin(pin)

            elif choice == 5:
                print("\nThank You for Using ATM")
                break

            else:
                print("Invalid Choice")
        break
    else:
        attempts -= 1
        print("Wrong PIN. Attempts left:", attempts)

        if attempts == 0:
            print("\nCARD BLOCKED. Contact Bank")
            break
```

# Real Benefits of Scope (Why Itâ€™s Important?)

```python

Increases Security
Sensitive data such as ATM PINs, API keys, and passwords remain protected when kept local, because they cannot be accessed from outside the function.

Saves Memory
Variables are stored only where they are needed, which helps the program use memory efficiently.

Reduces Bugs
Excessive use of global variables can cause confusion and errors.
Proper scope management results in cleaner and more stable code.

Makes Large Projects Manageable
Clearly defined scopes make maintaining, scaling, and organizing code much easier.

Keeps Functions Independent
Local variables ensure that each function behaves predictably without interfering with others.
```
