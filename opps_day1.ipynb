File Pointers in Python
In Python, a file pointer is a marker that indicates the current position in a file where the next read or write operation will occur. When you open a file, the file pointer is typically positioned at the beginning of the file. As you read from or write to the file, the file pointer moves accordingly.

```python
with open(r"C:\Users\Vishnu\Desktop\Python_Full Stack\File_handling.txt","r") as f:
    f.read()
    print(f.tell())
```

```python
with open(r"C:\Users\Vishnu\Desktop\Python_Full Stack\File_handling.txt","a+") as f:
   
    f.write("\n This is added using a+ mode.")
    f.seek(82)
    print(f.read())
```
