```python
a=int(input("Enter Your Age"))
if a>=18:
    print("Yes You Are Eligible")
else:
    print("Not Eligible")
```

```python
# using elif
student_input=int(input("Enter Your Marks"))
if student_input>=300:
    print("Your Grade is A")
elif student_input>=250:
    print("Your Grade is B")
elif student_input>=200:
    print("Your Grade is C")
else:
    print("Fail")
```

```python
pin=2134
balance=5000
enter_pin=int(input("Enter Your PIN"))
if enter_pin==pin:
    print("Login Success")
    amount=int(input("Enter Your Amount"))
    if amount<=balance:
        if amount%100==0:    
            if amount>0:
                print("Withdraw Success")
                balance-=amount
                print("Your Remaining Balance is ",balance)
            else:
                print("Please Entered Valid Amount")
        else:
            print("Please Enter Amount is Multiple of 100")
    else:
        print("Insufficient Amount")
else:
    print("Incorrect PIN")
```
